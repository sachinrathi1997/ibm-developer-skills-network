# Question 2: Extracting Tesla Revenue Data Using Webscraping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL of the webpage
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML data using BeautifulSoup with html.parser
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table containing Tesla Revenue
table = soup.find_all('table')[0]  # Assuming the first table is the one we want

# Initialize lists to store the data
dates = []
revenues = []

# Extract data from the table
for row in table.find_all('tr')[1:]:  # Skip the header row
    cols = row.find_all('td')
    date = cols[0].text.strip()
    revenue = cols[1].text.strip().replace('$', '').replace(',', '')  # Clean the revenue data
    dates.append(date)
    revenues.append(revenue)

# Create a DataFrame from the extracted data
tesla_revenue = pd.DataFrame({
    'Date': dates,
    'Revenue': revenues
})

# Display the last five rows
print(tesla_revenue.tail())
```

# Question 5: Tesla Stock and Revenue Dashboard

```python
import yfinance as yf
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# Function to make the graph
def make_graph(stock_data, revenue_data, title):
    # Ensure data up to June 2021
    stock_data_filtered = stock_data.loc[stock_data['Date'] <= '2021-06-30']
    
    # Convert 'Date' columns to datetime format
    stock_data_filtered.loc[:, 'Date'] = pd.to_datetime(stock_data_filtered['Date'])
    revenue_data.loc[:, 'Date'] = pd.to_datetime(revenue_data['Date'])
    revenue_data.loc[:, 'Revenue'] = pd.to_numeric(revenue_data['Revenue'])

    # Plot stock data
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data_filtered['Date'], stock_data_filtered['Close'], label='Closing Price', color='blue')
    
    # Plot revenue data
    plt.plot(revenue_data['Date'], revenue_data['Revenue'], label='Revenue', color='red', linestyle='--')

    # Add title and labels
    plt.title(f'{title} Stock Data Up to June 2021')
    plt.xlabel('Date')
    plt.ylabel('Value (USD)')
    plt.legend()
    plt.grid(True)
    
    # Display the graph
    plt.show()

# Extracting Tesla Stock Data Using yfinance
tesla_ticker = yf.Ticker("TSLA")
tesla_data = tesla_ticker.history(period="max")
tesla_data.reset_index(inplace=True)
print(tesla_data.head())

# Extracting Tesla Revenue Data Using Webscraping
tesla_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
tesla_response = requests.get(tesla_url)
tesla_soup = BeautifulSoup(tesla_response.text, 'html.parser')

# Find the table containing Tesla Revenue
tesla_table = tesla_soup.find_all('table')[0]  # Assuming the first table is the one we want

# Initialize lists to store the data
tesla_dates = []
tesla_revenues = []

# Extract data from the table
for row in tesla_table.find_all('tr')[1:]:  # Skip the header row
    cols = row.find_all('td')
    date = cols[0].text.strip()
    revenue = cols[1].text.strip().replace('$', '').replace(',', '')  # Clean the revenue data
    tesla_dates.append(date)
    tesla_revenues.append(revenue)

# Create a DataFrame from the extracted data
tesla_revenue = pd.DataFrame({
    'Date': tesla_dates,
    'Revenue': tesla_revenues
})

print(tesla_revenue.tail())

# Use the make_graph function to plot Tesla data
make_graph(tesla_data, tesla_revenue, 'Tesla')
```
